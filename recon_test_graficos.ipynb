```python
# 100k

import numpy as np
import matplotlib.pyplot as plt

group1 = [1.044, 0.199, 0.811, 0.704, 0.634, 0.128]
group2 = [89.987, 89.730, 89.716, 89.997, 88.964, 79.041]
group3 = [38.929, 32.312, 28.573, 22.710, 16.196, 8.162]
group4 = [35.018, 26.477, 21.982, 14.519, 10.624, 4.632]

x_labels = [11, 15, 20, 27, 35, 38]
x = range(len(x_labels))  # posiciones para los ticks

# Graficar
plt.figure(figsize=(8, 5))

plt.plot(x, group1, marker='o', label='Min')
plt.plot(x, group2, marker='s', label='Max')
plt.plot(x, group3, marker='^', label='Mean')
plt.plot(x, group4, marker='x', label='Median')

plt.xticks(ticks=x, labels=x_labels)
plt.xlabel('Cantidad de líneas')
plt.ylabel('Valor')
plt.title('Angulo min, max, mean, median para 100k')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
